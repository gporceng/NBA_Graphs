```python
from nba_api.stats.endpoints import playercareerstats

# Nikola Jokić
career = playercareerstats.PlayerCareerStats(player_id='203999') 

# pandas data frames (optional: pip install pandas)
career.get_data_frames()[0]

# json
career.get_json()

# dictionary
career.get_dict()
```

```python
#Jokic players stats
career.get_data_frames()[0]
```

```python
career.get_data_frames()[1]
```

```python
df = career.get_data_frames()
type(df[1])
```

```python
df[0]
# scatter plot
df[0].plot(kind = 'line',x = 'SEASON_ID',y = 'PTS',color = 'red',title='Points/Year Nikola Jokic')
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots()

# plot the first y-value
df[0].plot(kind='line', x='SEASON_ID', y='PTS', color='red', ax=ax1)
ax1.set_xlabel('SEASON_ID')
ax1.set_ylabel('Points')

# create a second y-axis and plot the second y-value
ax2 = ax1.twinx()
df[0].plot(kind='line', x='SEASON_ID', y='REB', color='blue', ax=ax2)
ax2.set_ylabel('Rebounds')

plt.title('Points/Year Nikola Jokic')
plt.show()
```

```python
import matplotlib.pyplot as plt

# create a larger figure size
fig, ax1 = plt.subplots(figsize=(8, 5))

# plot the first y-value
df[0].plot(kind='line', x='SEASON_ID', y='PTS', color='red', ax=ax1)
ax1.set_xlabel('Season', fontsize=12)
ax1.set_ylabel('Points', fontsize=12)

# create a second y-axis and plot the second y-value
ax2 = ax1.twinx()
df[0].plot(kind='line', x='SEASON_ID', y='REB', color='blue', ax=ax2)
ax2.set_ylabel('Rebounds', fontsize=12)

plt.title('Points/Year Nikola Jokic', fontsize=14)
plt.show()
```
